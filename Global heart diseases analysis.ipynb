# Data Reading


```python
pd.read_csv('/content/Dataset Heart Disease.csv')
pd.read_csv('/content/cleveland1.csv')
pd.read_csv('/content/cleveland2.csv')



df1 = pd.read_csv('/content/Dataset Heart Disease.csv')
df2 = pd.read_csv('/content/cleveland1.csv')
df3 = pd.read_csv('/content/cleveland2.csv')

df = pd.concat([df1, df2, df3], ignore_index=True)
```

# Cleaning (Caping)

```python

df['sex'] = df['sex'].replace({1: 'male', 0: 'female'})

for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)
df = df[df['cholesterol'] != 0]

if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])
    print("Dropped unnamed column successfully.")
else:
    print("No unnamed column found.")
```

```python

sns.set_style("whitegrid")

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

sns.histplot(df['age'], bins=20, kde=True, color='skyblue', ax=axes[0, 0])
axes[0, 0].set_title('Age Distribution')
axes[0, 0].set_xlabel('Age')
axes[0, 0].set_ylabel('Frequency')

sns.histplot(df['cholesterol'], bins=20, kde=True, color='lightgreen', ax=axes[0, 1])
axes[0, 1].set_title('Cholesterol Distribution')
axes[0, 1].set_xlabel('Cholesterol')
axes[0, 1].set_ylabel('Frequency')

sns.pairplot(df, vars=['age', 'cholesterol', 'resting bps', 'max heart rate'], hue='sex', palette='husl')
plt.suptitle('Pair Plot of Selected Features', y=1.02)

sns.scatterplot(x='age', y='cholesterol', hue='sex', data=df, ax=axes[1, 0], palette='coolwarm')
axes[1, 0].set_title('Cholesterol vs. Age by Sex')
axes[1, 0].set_xlabel('Age')
axes[1, 0].set_ylabel('Cholesterol')

fig.delaxes(axes[1, 1])

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python

#--------------------------------------------------#
sns.set_style("whitegrid")

plt.figure(figsize=(20, 14))

plt.subplot(2, 2, 1)
sns.histplot(data=df, x='age', hue='sex', multiple='stack', palette='pastel', bins=20)
plt.title('Age Distribution by Sex')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.subplot(2, 2, 2)
sns.countplot(data=df, x='chest pain type', palette='viridis')
plt.title('Chest Pain Type Distribution')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')

plt.subplot(2, 2, 3)
sns.scatterplot(data=df, x='cholesterol', y='max heart rate', hue='sex', palette='cool', alpha=0.7)
plt.title('Cholesterol vs. Max Heart Rate')
plt.xlabel('Cholesterol')
plt.ylabel('Max Heart Rate')

plt.subplot(2, 2, 4)

sns.countplot(x='sex', hue='target', data=df, palette='pastel')
plt.title('Heart Disease Prevalence by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()

plt.tight_layout()

plt.show()
```
